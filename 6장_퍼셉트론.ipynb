### AND 게이트 구현

```
import numpy as np

def AND(x1, x2, w1 = 0.5, w2 = 0.5, b = -0.7):
  x = np.array([x1, x2])
  w = np.array([w1, w2])
  y = np.sum(x * w) + b
  
  if y >= 0:
    return 1
  elif y < 0:
    return 0
```

```
print('AND(0,0) = ', AND(0,0))
print('AND(0,1) = ', AND(0,1))
print('AND(1,0) = ', AND(1,0))
print('AND(1,1) = ', AND(1,1))
```

### NAND 게이트 구현

```
import numpy as np

def NAND(x1, x2, w1 = -0.5, w2 = -0.5, b = 0.7):
  x = np.array([x1, x2])
  w = np.array([w1, w2])
  y = np.sum(x * w) + b
  
  if y >= 0:
    return 1
  elif y < 0:
    return 0
```

```
print('NAND(0,0) = ', NAND(0,0))
print('NAND(0,1) = ', NAND(0,1))
print('NAND(1,0) = ', NAND(1,0))
print('NAND(1,1) = ', NAND(1,1))    # NAND 게이트는 AND 게이트의 부정 -> 결과 반대
```

### OR 게이트 구현

```
import numpy as np

def OR(x1, x2, w1 = 0.5, w2 = 0.5, b = -0.2):
  x = np.array([x1, x2])
  w = np.array([w1, w2])
  y = np.sum(x * w) + b
  
  if y <= 0:
    return 0
  elif y > 0:
    return 1
```

```
print('OR(0,0) = ', OR(0,0))
print('OR(0,1) = ', OR(0,1))
print('OR(1,0) = ', OR(1,0))
print('OR(1,1) = ', OR(1,1))
```

### XOR 게이트 구현

```
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)
  return y
```

```
print('XOR(0,0) = ', XOR(0,0))
print('XOR(0,1) = ', XOR(0,1))
print('XOR(1,0) = ', XOR(1,0))
print('XOR(1,1) = ', XOR(1,1))
```

### 코딩으로 XOR 문제 해결하기

```
import numpy as np

# 가중치와 바이어스

w11 = np.array([-2, -2])
w12 = np.array([2, 2])
w2 = np.array([1, 1])
b1 = 3
b2 = -1
b3 = -1
```

```
# 퍼셉트론
def MLP(x, w, b):   # MLP(Multi Layer Perceptron)
  y = np.sum(w * x) + b
  if y <= 0:
    return 0
  else:
    return 1

# NAND gate
def NAND(x1, x2):
  return MLP(np.array([x1, x2]), w11, b1)

# OR gate
def OR(x1, x2):
  return MLP(np.array([x1, x2]), w12, b2)

# AND gate
def AND(x1, x2):
  return MLP(np.array([x1, x2]), w2, b3)

# XOR gate
def XOR(x1, x2):
  return AND(NAND(x1, x2), OR(x1, x2))
```

```
# x1, x2 값을 번갈아 대입해 가며 최종값 출력

if __name__ == '__main__':
  for x in [(0,0), (1,0), (0,1), (1,1)]:
    y = XOR(x[0], x[1])
    print("입력 값: " + str(x) + " 출력 값: " + str(y))
```
