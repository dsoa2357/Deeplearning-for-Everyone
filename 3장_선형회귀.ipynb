```
import numpy as np
```

```
x = [2, 4, 6, 8]
y = [81, 93, 91, 97]
```

```
mx = np.mean(x)
my = np.mean(y)

print("x의 평균값: ", mx)
print("y의 평균값: ", my)
```

```
divisor = sum([(mx - i)**2 for i in x])  # 분모
```

```
def top(x, mx, y, my):
  d = 0
  for i in range(len(x)):
    d += (x[i] - mx) * (y[i] - my)
  return d
dividend = top(x, mx, y, my)    # 분자
```

```
print("분모: ", divisor)
print("분자: ", dividend)
```

```
a = dividend / divisor
b = my - (mx * a)
```

```
print("기울기 a: ", a)
print("y 절편 b: ", b)
```

### 평균 제곱근 오차 구하기

```
import numpy as np
```

```
ab = [3, 76]  # 임의의 값
```

```
data = [[2,81], [4,93], [6,91], [8,97]]
x = [i[0] for i in data]
y = [i[1] for i in data]
```

```
def predict(x):
  return ab[0] * x + ab[1]
```

```
# rmse 함수

def rmse(p, a):
  return np.sqrt(((p-a)**2).mean())
```

```
# 최종값을 구하는 함수

def rmse_val(predict_result, y):
  return rmse(np.array(predict_result), np.array(y))
```

```
# 예측값 들어갈 빈 리스트

predict_result = []

for i in range(len(x)):
  predict_result.append(predict(x[i]))
  print("공부한 시간 = %.f, 실제 점수 = %.f, 예측 점수 = %.f" %(x[i], y[i], predict(x[i])))
```

```
# 최종 rmse 값 출력

print("rmse 최종값 : " + str(rmse_val(predict_result, y)))
```
