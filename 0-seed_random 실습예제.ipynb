### seed 설정
- 어떤 특정한 시작 숫자를 정해주면 정해진 알고리즘에 의해 마치 난수처럼 보이는 수열을 생성한다.
- 이런 시작 숫자를 시드(seed)라 한다.
- 시드 값은 한 번만 정해 주면 된다.

```
import numpy as np

# 시드 설정 1)
np.random.seed(0)
```

```
# 시드 설정 2)

seed=0
np.random.seed(seed)
```

```
np.random.seed(0) # 0이 아니고 임의의 숫자만 넣어도 됨
np.random.rand(5) # 위 함수가 없으면 값이 계속 변경
```

```
np.random.seed(100)
np.random.rand(10)
```

```
# 난수와 비슷하지만 시드 값에 따라 항상 같은 결과를 반환

np.random.seed(76923)

a = np.random.random((2,2))
b = np.random.sample((2,2))
c = np.random.randint(1, 3, (2,2), dtype=int)

print(a)
print()
print(b)
print()
print(c)
```

### 난수 발생
- numpy의 random은 난수를 생성하는 다양한 명령을 제공
- 가장 간단하고 많이 사용되는 것
  - rand: 0~1 사이의 균일 분포, 실수 난수
  - randn: 가우시안 표준 정규분포, 실수 난수
  - randint: 균일분포의 정수난수

### 1) rand명령
- 0부터 1 사이의 균일한 확률 분포로 실수 난수를 생성
- 숫자 인수는 생성할 난수의 크기
- 여러 개의 인수를 넣으면 해당 크기를 가진 행렬을 생성

```
np.random.rand(10)
```

```
np.random.rand(3,5)
```

### 2) randn
- 기댓값이 0이고 표준편차가 1인 가우시안 표준 정규 분포를 따르는 난수를 생성

```
np.random.randn(5)
```

```
np.random.randn(10)
```

```
np.random.randn(3,5)
```

### 3) randint
- randint(low, high=None, size=None)

```
np.random.randint(10, size=10)
```

```
np.random.randint(10, 20, size=10)
```

```
np.random.randint(10, 20, size=(3,5))
```
