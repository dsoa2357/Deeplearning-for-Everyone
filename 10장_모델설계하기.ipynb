### 폐암 수술 환자의 생존율 예측하기

```
from keras.models import Sequential
from keras.layers import Dense

import numpy as np
import tensorflow as tf
```

```
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)    # tf.set_random_seed(seed) 교재에 있는 것은 버전 업 되면서 오류
```

```
# delimiter : 세 자리마다 ',' 사용
Data_set = np.loadtxt("/content/gdrive/My Drive/deepLearning/data/ThoraricSurgery.csv", delimiter=",")
```

```
X = Data_set[:, 0:17] # 17개의 속성
Y = Data_set[:, 17]   # 생존 여부
```

```
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu'))   # 은닉층
model.add(Dense(1, activation='sigmoid'))               # 출력층
```

```
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(X, Y, epochs=30, batch_size=10)
print("\n Accuracy: %.4f" %(model.evaluate(X, Y)[1]))
```
